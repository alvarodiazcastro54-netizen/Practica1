```python
import numpy as np
import matplotlib.pyplot as plt

# Sistema de ecuaciones
a = np.array([[5,2],[-3,3]])
b = np.array([1,5])
sol = np.linalg.solve(a,b)
x_sol, y_sol = sol

# Rango de x para graficar las lÃ­neas
x = np.linspace(-2, 2, 400)

# Ecuaciones despejadas para y
y1 = (1 - 5*x)/2
y2 = (5 + 3*x)/3

plt.figure(figsize=(8,6))
plt.plot(x, y1, label='5x + 2y = 1')
plt.plot(x, y2, label='-3x + 3y = 5')
plt.plot(x_sol, y_sol, 'ro', label=f"SoluciÃ³n: x = {x_sol}, y = {y_sol}")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sistema de ecuaciones lineales')
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio nÂº 11

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir variables
x, y = sp.symbols('x y')

# Definir ecuaciones
eq1 = sp.Eq(2*(x + 4)/3 - y/2, 9/2)
eq2 = sp.Eq(x + 2*y - (1/3)*(3*x - 2), -4/3)

# Resolver sistema
sol = sp.solve((eq1, eq2), (x, y))
print(f"SoluciÃ³n exacta: x = {sol[x]}, y = {sol[y]}")

# Despejar y en ambas ecuaciones para graficar
y_expr1_solved = sp.solve(eq1, y)[0]
y_expr2_solved = sp.solve(eq2, y)[0]

# Convertir expresiones simbÃ³licas a funciones numÃ©ricas
y_func1 = sp.lambdify(x, y_expr1_solved, "numpy")

# Crear rango de values para x
x_vals = np.linspace(-10, 10, 400)

# Graficar
plt.figure(figsize=(8,6))

# Plot the first equation
y_vals1 = y_func1(x_vals)
plt.plot(x_vals, y_vals1, label=f'EcuaciÃ³n 1: {eq1}')


# Check if y_expr2_solved depends on x before lambdifying and plotting
if y_expr2_solved.has(x):
    y_func2 = sp.lambdify(x, y_expr2_solved, "numpy")
    y_vals2 = y_func2(x_vals)
    plt.plot(x_vals, y_vals2, label=f'EcuaciÃ³n 2: {eq2}')
else:
    # If y_expr2_solved does not depend on x, it's a horizontal line at the solved y value
    plt.axhline(y=float(sol[y]), color='orange', linestyle='--', label=f'EcuaciÃ³n 2: {eq2}')


# Plot the solution point
plt.plot(float(sol[x]), float(sol[y]), 'ro', label='SoluciÃ³n')

plt.xlabel('x')
plt.ylabel('y')
plt.title('GrÃ¡fica de sistema con fracciones usando SymPy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir variables
x, y = sp.symbols('x y')

# Definir la ecuaciÃ³n
eq = sp.Eq(5*x - 4*y, 1)

# a) Buscar dos pares de valores que solucionan la ecuaciÃ³n
x_values = [0, 1]
solutions = []
for val in x_values:
    sol = sp.solve(eq.subs(x, val), y)
    solutions.append((val, sol[0]))

print("Dos pares de soluciones:")
for i, (x_val, y_val) in enumerate(solutions, 1):
    print(f"SoluciÃ³n {i}: x = {x_val}, y = {y_val}")

# b) Graficar la recta 5x - 4y = 1
# Despejar y para graficar: 5x -4y = 1 => y = (5x - 1)/4
y_expr = (5*x - 1)/4

# Convertir expresiÃ³n simbÃ³lica a funciÃ³n numÃ©rica
y_func = sp.lambdify(x, y_expr, "numpy")

# Crear rango de valores para x
x_vals = np.linspace(-10, 10, 400)
y_vals = y_func(x_vals)

# Graficar
plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label='5x - 4y = 1')

# Graficar los puntos soluciÃ³n encontrados
for (x_sol, y_sol) in solutions:
    plt.plot(x_sol, float(y_sol), 'ro')
    plt.text(x_sol, float(y_sol), f'({x_sol},{y_sol})')

plt.xlabel('x')
plt.ylabel('y')
plt.title('GrÃ¡fica de la recta 5x - 4y = 1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Volver a definir variables y ecuaciÃ³n por si no estÃ¡n en memoria
x, y = sp.symbols('x y')
eq = sp.Eq(3*x - 2*y, 1)

# Despejar y: 3x - 2y = 1 â y = (3x - 1)/2
y_expr = sp.solve(eq, y)[0]
y_func = sp.lambdify(x, y_expr, 'numpy')

# Rango para x
x_vals = np.linspace(-10, 10, 400)
y_vals = y_func(x_vals)

# Usar los mismos puntos obtenidos en el apartado a)
puntos = [(0, sp.solve(eq.subs(x, 0), y)[0]),
          (2, sp.solve(eq.subs(x, 2), y)[0])]

# Graficar
plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label='Recta: 3x - 2y = 1', color='blue')

# Graficar los puntos en rojo
for (x_val, y_val) in puntos:
    plt.plot(x_val, float(y_val), 'ro')
    plt.text(x_val + 0.2, float(y_val), f'({x_val},{y_val})', fontsize=9)

plt.xlabel('x')
plt.ylabel('y')
plt.title('GrÃ¡fica de la recta 3x - 2y = 1')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # eje x
plt.axvline(0, color='black', linewidth=0.5)  # eje y
plt.legend()
plt.show()
```

```python
import sympy as sp

# Definir variables y ecuaciÃ³n
x, y = sp.symbols('x y')
eq = sp.Eq(5*x + 2*y, 3)

# La ecuaciÃ³n tiene infinitas soluciones (es una recta)
print("b) La ecuaciÃ³n tiene infinitas soluciones, porque representa una recta.")

# Obtener 2 soluciones eligiendo valores de x
soluciones = []
for x_val in [0, 1]:
    y_val = sp.solve(eq.subs(x, x_val), y)[0]
    soluciones.append((x_val, y_val))

# Mostrar soluciones
for i, (x_val, y_val) in enumerate(soluciones, 1):
    print(f"SoluciÃ³n {i}: x = {x_val}, y = {y_val}")
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definir variables simbÃ³licas
x, y = sp.symbols('x y')

# Definir la ecuaciÃ³n x + 2y = 4
eq = sp.Eq(x + 2*y, 4)

# Despejar y
y_expr = sp.solve(eq, y)[0]  # y = (4 - x)/2

# Convertir a funciÃ³n numÃ©rica
y_func = sp.lambdify(x, y_expr, 'numpy')

# Rango de valores para graficar
x_vals = np.linspace(-2, 6, 400)
y_vals = y_func(x_vals)

# Puntos que pertenecen a la recta
points = [(0, 2), (2, 1), (4, 0)]

# Graficar
plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label='x + 2y = 4', color='blue')
plt.scatter(*zip(*points), color='red', label='Puntos soluciÃ³n')

# Anotar los puntos
for (x_val, y_val) in points:
    plt.annotate(f'({x_val},{y_val})', (x_val, y_val), textcoords="offset points", xytext=(5,5), ha='left')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Recta x + 2y = 4 usando SymPy y Matplotlib')
plt.grid(True)
plt.legend()
plt.show()
```
